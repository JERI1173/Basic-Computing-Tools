```python
# Problem 1

# part A

import numpy as np
from scipy.linalg import cholesky, solve_triangular

# Cholesky decomposition
def solve_chol(A, b):
    """Solve a linear algebra problem with Cholesky decomposition

    Parameters:
    - A: ndarray, SPD matrix
    - b: ndarray, vector

    Return
    - x: ndarray, vector
    """
    # Obtain L
    L = cholesky(A, lower = True) # A = L @ L.T
    
    # Obtain y 
    y = solve_triangular(L, b, lower=True) # L @ y = b

    # Obtain x 
    x = solve_triangular(L.T, y) # L.T @ x = y
    
    return x
```

```python
# part B

from scipy.linalg import cholesky, lu_factor
import time

# Generate logarithmically spaced values for n
time_span = np.round(np.logspace(1, np.log10(2000), 10)).astype(int)

cholesky_times = []
LU_times = []

for n in time_span:
    # Generate n * n SPD matrix
    matrix = np.random.randn(n, n)
    matrix = matrix.T @ matrix
    
    # Cholesky decomposition
    start_time = time.time()
    cholesky(matrix)
    cholesky_times.append(time.time() - start_time)
    
    # LU decomposition
    start_time = time.time()
    lu_factor(matrix)
    LU_times.append(time.time() - start_time)
    
# Plot
plt.loglog(time_span, cholesky_times,label='Cholesky decomposition', marker='o')
plt.loglog(time_span, LU_times,label='LU decomposition', marker='o')

# Title
plt.title('Cholesky vs. LU Decomposition Time')

# x, y lable
plt.xlabel('Matrix size(n)')
plt.ylabel('Time(sec)')

# Legend
plt.legend()
plt.show()
```

```python
# part C
from numpy.linalg import eigh

def matrix_pow(A, n):
    """Compute the power of a symmetric matrix with eigenvalue decomposition
    
    Parameters:
    - A: ndarray, symmetric matrix
    - n: int, power
    
    Return:
    - A_n: ndarray, the matrix A**n
    """
    # Compute eigenvalue decomposition of A
    L, Q = eigh(A) # A = Q @ L @ Q.T, L is a diagonal matrix

    # Compute L**n element-wise
    L_n = np.diag(L**n) # L is diagonal

    # Compute A**n using the eigenvalue decomposition
    A_n = Q @ L_n @ Q.T

    return A_n
```

```python
# part D

from scipy.linalg import lu

def abs_det(A):
    """Compute the determinant with LU decomposition
    
    Parameters:
    - A: ndarray, square matrix
    
    Return:
    - abs(det_A): int, absolute value of determinant of matrix A
    """
    P, L, U = lu(A) # A = P @ L @ U
    det_A = np.prod(np.diag(U)) # det(A) = det(P) * det(L) * det(U)
    
    return abs(det_A)
```
